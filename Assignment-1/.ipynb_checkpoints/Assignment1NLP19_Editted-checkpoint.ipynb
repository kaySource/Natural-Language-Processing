### Task A: In this sub-task, you are expected to carry out the following tasks:

**Tokenize** the corpus into sentences and words (for each of the pos and neg class). **Print the number of sentences and words.** <br>
**Perform case-folding** on the corpus. <br>
**Remove the stopwords** from the corpus and print the count of the rest of the non stop-words occurring in the corpus.<br>

```python
#Write the code for Task A
import os
import re
import nltk
from nltk.corpus import stopwords 
from nltk.tokenize import sent_tokenize
from nltk.tokenize import RegexpTokenizer

files = []
for file in os.listdir('NCBI_Data'):
        if (file[-4:] == ".txt"):
            files.append('NCBI_Data/' + file)
# print (files)

corpus = ""
sentences = []
for file in files:
    with open(file) as f:
        file_txt = f.read().lower().strip()
        file_txt = re.sub('[ ]*-[ -]*', '-' , file_txt)
        file_txt = re.sub('[ ]*\'[ \']*', '\'', file_txt) 
        sentence_file = sent_tokenize(file_txt)
        # print(len(sentence_file))
        sentences.extend(sentence_file)
        corpus = corpus + file_txt + ".\n"
# print(corpus)
# print(sentences)

word_tokenizer = RegexpTokenizer('[\w]+[-\'][\w]+|[\w]+')
all_words = word_tokenizer.tokenize(corpus)
# print(all_words)
stop_words = set(stopwords.words('english'))
# print(stop_words)
word_tokens = [word for word in all_words if word not in stop_words]
# print(word_tokens)

print("Number of Sentences: ", len(sentences))
print("Total Number of Words: ", len(all_words))
print("Number of Word Tokens: ", len(word_tokens))
```

### Task B: In this sub-task, you are expected to carry out the following tasks:

1. **Create the following language models** on the training corpus: <br>
    i.   Unigram <br>
    ii.  Bigram <br>
    iii. Trigram <br>
    iv.  Fourgram <br>

2. **List the top 5 bigrams, trigrams, four-grams (with and without Add-1 smoothing).**
(Note: Please remove those which contain only articles, prepositions, determiners. For Example: “of the”, “in a”, etc).

```python
#Write the code for Task B

from nltk.util import ngrams
unigrams=[]
bigrams=[]
trigrams=[]
fourgrams=[] 

for sentence in sentences:
    unigrams.extend(ngrams(word_tokenizer.tokenize(sentence), 1))
    bigrams.extend(ngrams(word_tokenizer.tokenize(sentence), 2))
    trigrams.extend(ngrams(word_tokenizer.tokenize(sentence), 3))
    fourgrams.extend(ngrams(word_tokenizer.tokenize(sentence), 4))
```

```python
def isTupleInformative(tuple1):
    for word in tuple1:
        if word not in stop_words:
            return True
    return False

print("Without Add-1 Smoothing: \n")

uni_processed = [tuple1 for tuple1 in unigrams if isTupleInformative(tuple1)]
uni_fdist = nltk.FreqDist(uni_processed)
print("Top 10 Frequent Unigrams out of", len(uni_processed), ":", uni_fdist.most_common(10))
print("")

bi_processed = [tuple1 for tuple1 in bigrams if isTupleInformative(tuple1)]
bi_fdist = nltk.FreqDist(bi_processed)
print("Top 10 Frequent Bigrams out of", len(bi_processed), ":", bi_fdist.most_common(10))
print("")

tri_processed = [tuple1 for tuple1 in trigrams if isTupleInformative(tuple1)]
tri_fdist = nltk.FreqDist(tri_processed)
print("Top 10 Frequent Trigrams out of", len(tri_processed), ":" , tri_fdist.most_common(10))
print("")

four_processed = [tuple1 for tuple1 in fourgrams if isTupleInformative(tuple1)]
four_fdist = nltk.FreqDist(four_processed)
print("Top 10 Frequent Fourgrams out of", len(four_processed), ":", four_fdist.most_common(10))
print("")
```

### With Smoothing

```python
#You are to perform Add-1 smoothing here:

print("With Add-1 Smoothing: \n")

uni_fdist_1 = {}
for tuple1 in set(uni_processed):
    uni_fdist_1[tuple1] = (1 + uni_fdist[tuple1])/( len(set(uni_processed)) + len(uni_processed) )
most_frequent_uni = sorted(uni_fdist_1.items(), key = lambda x : x[1], reverse = True)
print("Top 10 Frequent Unigrams out of", len(uni_processed), ":", most_frequent_uni[:10])
print("")

bi_fdist_1 = {}
for tuple1 in set(bi_processed):
    bi_fdist_1[tuple1] = (1 + bi_fdist[tuple1])/( len(set(bi_processed)) + len(bi_processed) )
most_frequent_bi = sorted(bi_fdist_1.items(), key = lambda x : x[1], reverse = True)
print("Top 10 Frequent Bigrams out of", len(bi_processed), ":", most_frequent_bi[:10])
print("")
  
tri_fdist_1 = {}
for tuple1 in set(tri_processed):
    tri_fdist_1[tuple1] = (1 + tri_fdist[tuple1])/( len(set(tri_processed)) + len(tri_processed) )
most_frequent_tri = sorted(tri_fdist_1.items(), key = lambda x : x[1], reverse = True)
print("Top 10 Frequent Trigrams out of", len(tri_processed), ":", most_frequent_tri[:10])
print("")
  
four_fdist_1 = {}
for tuple1 in set(four_processed):
    four_fdist_1[tuple1] = (1 + four_fdist[tuple1])/( len(set(four_processed)) + len(four_processed) )
most_frequent_four = sorted(four_fdist_1.items(), key = lambda x : x[1], reverse = True)
print("Top 10 Frequent Fourgrams out of", len(four_processed), ":", most_frequent_four[:10])
print("")
  
```
